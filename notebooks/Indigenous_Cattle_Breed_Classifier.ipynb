# Create DataBunch

Now we have preprocessed data and we are ready for create databuch, so that we can feed them into our deep learning model

```
# transform data
tfms = get_transforms(max_rotate=20, max_zoom=1.3, max_lighting=0.4, 
                      max_warp=0.4, p_affine=1., p_lighting=1.)
```

```
# Create data bunch from folder
np.random.seed(42)
data = ImageDataBunch.from_folder(path, train=".", valid_pct=0.15, ds_tfms=tfms, size=224, num_workers=4).normalize(imagenet_stats)
```

# Modelling

For this project, I am using Convolutional Neural Network(CNN) model.

```
# Define the model
learner = create_cnn(data, models.resnet50, metrics = error_rate).mixup()
```

# Model Training

```
# train that model
learner.fit_one_cycle(4, max_lr=0.001)
```

So in just 4 epochs we are able to reduce error rate by 10%.

```
learner.fit_one_cycle(10)
```

```
learner.lr_find()
learner.recorder.plot()
```

```
learner.unfreeze()
```

```
learner.fit_one_cycle(50, slice(1e-7,1e-2))
```

```
# model was still underfitting, so lets trained it more say for 20 epochs
learner.fit_one_cycle(20, slice(1e-9,1e-2))
```

```
learner.fit_one_cycle(20, slice(1e-6,1e-2))
```

```
learner.recorder.plot_losses()
```

```
interp = ClassificationInterpretation.from_learner(learner)
```

```
interp.plot_top_losses(9, figsize=(15,11))
```

```
interp.most_confused(min_val=2)
```
